# 模块一：NumPy 核心概念与实践

欢迎来到数据分析之旅的第一站！NumPy (Numerical Python) 是 Python 科学计算的基石。它提供了强大的多维数组对象 `ndarray` 和丰富的函数库。

---

## 1. `ndarray` 对象：NumPy 的核心

`ndarray` 是一个高效的多维数组，用于存储和处理大规模数据。

### 1.1 `ndarray` 的两大特性

- **多维性 (N-dimensional)**：可以表示从 0 维（标量）到任意高维的数据。
- **同质性 (Homogeneous)**：数组中的所有元素必须是相同的数据类型。如果不同，NumPy 会自动进行向上转换（如 `int` -> `float`, `float` -> `string`）。

```python
import numpy as np

# 0维数组
arr_0d = np.array(5)
print(f"0维数组: {arr_0d}, 维度: {arr_0d.ndim}")

# 1维数组
arr_1d = np.array([1, 2, 3])
print(f"\n1维数组: \n{arr_1d}, \n维度: {arr_1d.ndim}")

# 2维数组
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(f"\n2维数组: \n{arr_2d}, \n维度: {arr_2d.ndim}")
```

```python
# 演示同质性：int 和 str -> str
arr_str = np.array([1, 2, "hello"])
print(f"\n混合 str 后的数组: {arr_str}, 数据类型: {arr_str.dtype}")

# 演示同质性：int 和 float -> float
arr_float = np.array([1, 2.5])
print(f"混合 float 后的数组: {arr_float}, 数据类型: {arr_float.dtype}")
```

### 1.2 `ndarray` 的常用属性

了解数组的属性是进行操作的基础。

- `.ndim`: 维度数量
- `.shape`: 数组的形状（各维度的长度）
- `.size`: 元素总数
- `.dtype`: 元素的数据类型
- `.T`: 数组的转置（交换行和列）

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(f"数组: \n{arr}")
print(f"形状 (shape): {arr.shape}")
print(f"维度 (ndim): {arr.ndim}")
print(f"元素总数 (size): {arr.size}")
print(f"数据类型 (dtype): {arr.dtype}")
print(f"转置 (T): \n{arr.T}")
```

### 1.3 `ndarray` 的创建方法

除了从列表创建，NumPy 还提供了多种高效的数组创建方式。

```python
# --- 基础创建 ---
arr_from_list = np.array([4, 5, 6], dtype=np.float64)
print(f"从列表创建: {arr_from_list}")

# 深拷贝
arr_copy = np.copy(arr_from_list)
arr_copy[0] = 99
print(f"拷贝后的数组: {arr_copy}")
print(f"原始数组不变: {arr_from_list}")

# --- 预定义形状 ---
print(f"\n全零数组 (2,3):\n{np.zeros((2, 3), dtype=int)}")
print(f"全一数组 (2,3):\n{np.ones((2, 3))}")
print(f"未初始化数组 (2,3):\n{np.empty((2, 3))}") # 值是内存中的任意值
print(f"固定值数组 (2,3), 填充 2025:\n{np.full((2, 3), 2025)}")
print(f"形状与 arr 相同，但填充1:\n{np.ones_like(arr)}")

# --- 等差序列 ---
print(f"\narange(2, 11, step=2): {np.arange(2, 11, 2)}")
print(f"linspace(0, 10, 5个点): {np.linspace(0, 10, 5)}") # 包含终点
print(f"logspace(0, 4, 3个点, base=2): {np.logspace(0, 4, 3, base=2)}") # 2^0, 2^2, 2^4

# --- 特殊矩阵 ---
print(f"\n单位矩阵 (3,4):\n{np.eye(3, 4, dtype=int)}")
print(f"对角矩阵:\n{np.diag([5, 1, 2, 3])}")

# --- 随机数组 ---
np.random.seed(42) # 设置随机种子以保证结果可复现
print(f"\n0-1均匀分布 (2,3):\n{np.random.rand(2, 3)}")
print(f"标准正态分布 (2,3):\n{np.random.randn(2, 3)}")
print(f"随机整数 [3, 30) (2,3):\n{np.random.randint(3, 30, (2, 3))}")
```

### 1.4 `ndarray` 的索引与切片

与 Python 列表类似，但功能更强大，尤其是在多维数组上。

```python
# --- 一维数组 ---
arr_1d = np.arange(10)
print(f"一维数组: {arr_1d}")
print(f"索引第3个元素: {arr_1d[2]}")
print(f"切片 [2:5]: {arr_1d[2:5]}")
# 布尔索引：获取所有大于5的元素
print(f"布尔索引 (arr > 5): {arr_1d[arr_1d > 5]}")

# --- 二维数组 ---
arr_2d = np.random.randint(1, 100, (4, 5))
print(f"\n二维数组:\n{arr_2d}")
# 索引第2行第4列的元素 (行和列索引都是从0开始)
print(f"索引 [1, 3]: {arr_2d[1, 3]}")
# 切片：获取第2行，从第3列到结尾
print(f"切片 [1, 2:]: {arr_2d[1, 2:]}")
# 布尔索引：获取所有大于50的元素
print(f"布尔索引 (arr > 50): {arr_2d[arr_2d > 50]}")
```

### 1.5 `ndarray` 的运算

NumPy 的核心优势在于其高效的向量化运算。

- **元素级运算 (Element-wise)**: 对数组中的每个元素执行相同操作。
- **广播机制 (Broadcasting)**: 在形状不同的数组间进行运算的强大机制。
- **矩阵运算**: 真正的线性代数矩阵乘法。

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# 元素级运算
print(f"a + b:\n{a + b}")
print(f"a * b:\n{a * b}")

# 广播机制 (一个二维数组和一个标量)
print(f"a + 10:\n{a + 10}")

# 矩阵乘法
print(f"a @ b (矩阵乘法):\n{a @ b}")
```

## 2. NumPy 常用函数

NumPy 提供了海量的数学和统计函数，是数据分析的利器。

### 2.1 统计函数

```python
data = np.random.randint(1, 100, 20)
print(f"数据: {data}")

print(f"总和: {np.sum(data)}")
print(f"平均值: {np.mean(data)}")
print(f"中位数: {np.median(data)}")
print(f"标准差: {np.std(data):.2f}")
print(f"方差: {np.var(data):.2f}")
print(f"最大值: {np.max(data)} (在索引 {np.argmax(data)})")
print(f"最小值: {np.min(data)} (在索引 {np.argmin(data)})")
print(f"80分位数: {np.percentile(data, 80)}") # 80%的数据小于该值
print(f"累积和: {np.cumsum(data)}")
```

### 2.2 条件与逻辑函数

```python
scores = np.random.randint(40, 101, 15)
print(f"学生成绩: {scores}")

# np.where(condition, x, y) 是 if-else 的向量化版本
grades = np.where(scores >= 60, '及格', '不及格')
print(f"是否及格: {grades}")

# np.select 用于多条件判断
conditions = [scores < 60, (scores >= 60) & (scores < 85), scores >= 85]
choices = ['不及格', '良好', '优秀']
detailed_grades = np.select(conditions, choices, default='未知')
print(f"详细等级: {detailed_grades}")
```

### 2.3 数组操作函数

```python
arr = np.random.randint(1, 10, (3, 4))
print(f"原始数组:\n{arr}")

# 排序
print(f"排序后的数组 (不改变原数组):\n{np.sort(arr, axis=1)}") # 按行排序

# 去重
arr_flat = arr.flatten() # 先展平为1维
print(f"展平后: {arr_flat}")
print(f"唯一值: {np.unique(arr_flat)}")

# 形状调整
print(f"调整形状为 (4, 3):\n{arr.reshape(4, 3)}")
```

## 3. NumPy 综合实战练习

让我们用刚才学到的知识解决一些实际问题。

```python
# 题目1：温度数据分析
temps = np.array([28, 38, 29, 31, 32, 30, 29])
print(f"平均气温: {np.mean(temps):.2f}°C")
print(f"最高气温: {np.max(temps)}°C")
print(f"气温超过30°C的天数: {np.sum(temps > 30)}")

# 题目4：随机数据生成与操作
np.random.seed(0)
data = np.random.randint(0, 10, (3, 4))
print(f"\n原始随机矩阵:\n{data}")
print(f"每列最大值: {np.max(data, axis=0)}")
print(f"每行最小值: {np.min(data, axis=1)}")
data_processed = np.where(data % 2 == 1, -1, data)
print(f"将奇数替换为-1:\n{data_processed}")

# 题目7：销售额统计
sales = np.array([120, 135, 110, 125, 130, 140])
print(f"\n销售额总和: {np.sum(sales)}")
print(f"销售额最高的月份 (索引+1): {np.argmax(sales) + 1}")
print(f"销售额最低的月份 (索引+1): {np.argmin(sales) + 1}")

# 题目9：唯一值与出现次数
arr = np.array([2, 1, 2, 3, 1, 4, 3])
unique_vals, counts = np.unique(arr, return_counts=True)
print(f"\n唯一值: {unique_vals}")
print(f"对应出现次数: {counts}")
```
